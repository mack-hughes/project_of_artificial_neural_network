```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import SGDRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neural_network import MLPClassifier
```

# Exercise 1

```python
df_heart_disease = pd.read_csv("/home/mack/Download/Datasets/heart-disease-uci/heart.csv")
df_heart_disease
```

```python
df_heart_disease.info()
```

```python
# Checking if there are any Nan value
df_heart_disease.isna().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_heart_disease.drop("target", axis=1), 
                                                    df_heart_disease["target"], 
                                                    random_state=42)
```

```python
clf = MLPClassifier(random_state=42, max_iter=300).fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
